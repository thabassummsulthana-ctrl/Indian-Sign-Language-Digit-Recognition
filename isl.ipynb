```python
import cv2
import os


BASE_FOLDER = "digits"   # main dataset folder
LABEL = "9"             # start with 0, then 1,2,...9
TOTAL_IMAGES = 150
IMG_SIZE = 224



save_path = os.path.join(BASE_FOLDER, LABEL)
os.makedirs(save_path, exist_ok=True)

cap = cv2.VideoCapture(0)
count = 0

print("Press S to save | Press Q to quit")
print(f"Collecting digit: {LABEL}")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.flip(frame, 1)

    # Green box
    x1, y1, x2, y2 = 100, 100, 400, 400
    cv2.rectangle(frame, (x1, y1), (x2, y2), (0,255,0), 2)

    hand = frame[y1:y2, x1:x2]
    hand = cv2.resize(hand, (IMG_SIZE, IMG_SIZE))

    cv2.imshow("Webcam", frame)
    cv2.imshow("Captured Hand", hand)

    key = cv2.waitKey(1)

    if key == ord('s'):
        cv2.imwrite(f"{save_path}/{count}.jpg", hand)
        count += 1
        print(f"Saved {count}/{TOTAL_IMAGES}")

    if count >= TOTAL_IMAGES:
        print("Done for this digit!")
        break

    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
